```python
model = ModelInference(
    model_id="mistralai/mistral-large",
    credentials=creds,
    project_id=PROJECT_ID
)
```

```python
prompt = """
You are a helpful travel planner AI. Create a detailed 5-day travel itinerary for a trip from Delhi to Bandikui.
Include:
- Hotel recommendations
- Transport options
- Sightseeing suggestions
- Daily schedule
"""

response = model.generate(
    prompt=prompt,
    params={
        "decoding_method": "greedy",
        "max_new_tokens": 500
    }
)
```
